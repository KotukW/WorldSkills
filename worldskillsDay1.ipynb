```python
dir_name = "/content/drive/MyDrive/set/" #Переменная, хранящая путь к файлам пациентов
files = os.listdir(dir_name) #Открытие папки с параметрами пациентов и считывание название каждого файла в папке
len(files) # Проверяем количество считанных названий файлов в папке
```

```python
patients_data = pd.DataFrame() #Датафрейм со всеми пациентами
for filename in files:
    patient_data = pd.read_csv(dir_name+filename) #Считываем файл
    new_index = patient_data.loc[0] #Берем первую строку из файла (ID пациента)
    patient_data = patient_data.drop([0]) #Удаляем первую строку из Дата Фрейма
    patient_data[new_index.Parameter] = new_index.Value #Добавляем первую строку - ID пациента, как столбец
    patient_data.index = patient_data.RecordID # Ставим этот столбец как индекс
    patient_data = patient_data.drop("RecordID", axis=1) #Удаляем ненужный столбец
    patients_data = patients_data.append(patient_data) #Добавляем в общий Дата Фрейм
```
